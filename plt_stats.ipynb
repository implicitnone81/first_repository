```python
%matplotlib inline
import glob
import matplotlib.pyplot as plt 
#import numpy as np

datadir = '/Users/fwrenn/cleaning/bad_scans/gms5_bad_scans/'
masklist=[]
for file in glob.glob(datadir+'*.bmp'):
    file = file.split('/')[-1]
    masklist.append(file)
    
# Create unique set of jpg files.
lst = []
for file in glob.glob(datadir+'*.jpg'):
    file = file.split('/')[-1]
    lst.append(file[0:31])

jpglist = []
for file in lst:
    if file not in jpglist:
        jpglist.append(file)
     
# Match name to mask name        
for i in range(0,len(jpglist)):
    jpglist[i] = jpglist[i]+'.mask.bmp'
# Find the differece of the two lists.
diff = sorted(list(set(jpglist)-set(masklist)))

# Create y
y = [len(jpglist)-len(diff),len(masklist)]

p1=plt.bar([1,2],y,align='center',alpha=0.4,width=0.5,label='Bad Scan Lines')
p2=plt.bar([1],len(diff),alpha=0.4,width=0.5,bottom=y[0],color='r',align='center')
plt.xticks([1,2],['Human Detected','Mask Detected'])
ax = plt.gca()
plt.legend([p1, p2], ["Common Detections", "Difference"])
plt.title('Human vs Mask Detected')
plt.ylabel('Hours Detected')
ax.set_xlim([0,3])
plt.show()
#plt.savefig('MET5.2000.03.ps',format='png')
```
